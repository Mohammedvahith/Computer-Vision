```python
import cv2
```

```python
#data visualization library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#image processing library
from skimage.feature import hog
from skimage import data, exposure
```

```python
img=cv2.imread('pic.jpg')
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

# HOG

```python
fd ,hog_image =hog(img1,orientations=9, pixels_per_cell=(5,5),cells_per_block=(20,20),visualize=True, channel_axis=2)
```

# Rescaling

```python
rescale = exposure.rescale_intensity(hog_image, in_range=(0,10))
```

# plot

```python
figure,(a1,a2)=plt.subplots(1,2,figsize=(12,6), sharex=True ,sharey=True)


a1.axis('off')
a1.imshow(img1)
a1.set_title('Input image')

a2.axis('off')
a2.imshow(rescale)
a2.set_title('Histogram of oriented gradients')
plt.show()
```
