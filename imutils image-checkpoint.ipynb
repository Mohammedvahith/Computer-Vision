```python
import cv2 
import imutils
import skimage  
     
# initializing the hog person
# detector
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# reading the image
image = cv2.imread('off.jpg')

# resizing the image
image = imutils.resize(image,width=min(400,image.shape[1]))

# Detectimg all the regions in the image that has a pedestrians inside it
(regions, img) = hog.detectMultiScale(image, winStride=(3,10),padding=(3,5),scale=1.05)

# Drawing the regions in the image 
for(x,y,w,h) in regions:
    cv2.rectangle(image,(x,y),(x + w,y + h),(0, 255, 0),2)
    
    
# Showimg the output Image
cv2.imshow("Image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
import matplotlib.pyplot as plt

# reading the image
img = imread('off.jpg')
plt.axis('off')
plt.imshow(img)
print(img.shape)

# resizing image 
resized_img = resize(img, (128*4,64*8))
plt.axis('off')
plt.imshow(resized_img)
print(resized_img.shape)
plt.show()

# creating hog features
fd, hog_image = hog(img, orientations=10, pixels_per_cell=(2,2),cells_per_block=(2,2),visualize=True, channel_axis=2)
print(fd.shape)
print(hog_image.shape)
plt.axis("on")
plt.imshow(hog_image, cmap="gray")
plt.show()
 
# save image 
plt.imsave("resized_img.jpg",img)
plt.imsave("hog_img.jpg",hog_image, cmap="gray")
```
