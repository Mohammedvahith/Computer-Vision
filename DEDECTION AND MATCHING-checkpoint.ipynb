```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

# Reading image 

```python
img = cv2.imread('vijay.jpg')
```

# colour conversion

```python
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

# SCALE INVARIANCE

```python
sc_iv = cv2.pyrDown(rgb)
sc_iv2 = cv2.pyrDown(sc_iv)
num_rows, num_cols = sc_iv2.shape[:2]
print(sc_iv2)
```

```python
cv2.imshow("original",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("first",sc_iv)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("second",sc_iv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# rotational invariance

```python
rot_iv = cv2.getRotationMatrix2D((num_cols/2,num_rows/2),0,0.5)
```

# AFFINE TRANSFORMATION

```python
aff = cv2.warpAffine(sc_iv,rot_iv,(num_cols ,num_rows))
```

# colour conversion

```python
rot_gray = cv2.cvtColor(aff, cv2.COLOR_RGB2GRAY)
```

# rotational display

```python
fx, plots = plt.subplots(1,2,figsize=(20,10))

plots[0].set_title("training image")
plots[0].imshow(rgb)

plots[1].set_title("testing image")
plots[1].imshow(aff)
```

# SIFT

```python
sift=cv2.xfeatures2d.SIFT_create()
```

# DETECT AND COMPUTE

```python
train_kp,train_des = sift.detectAndCompute(gray,None)
test_kp,test_des = sift.detectAndCompute(rot_gray,None)
```

# keypoints


```python
kp_wo_size=np.copy(rgb)
kp_w_size=np.copy(rgb)
```

# DRAWING KEYPOINTS

```python
cv2.drawKeypoints(rgb, train_kp, kp_wo_size, color = (255,0,0))
cv2.drawKeypoints(rgb, train_kp, kp_w_size, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

# KEYPOINTS DISPLAY


```python
fx, plots = plt.subplots(1,2,figsize=(20,10))

plots[0].set_title("training keypoint with size")
plots[0].imshow(kp_w_size, cmap ='gray')

plots[1].set_title("training keypoint without size")
plots[1].imshow(kp_wo_size, cmap ='gray')
```

# BRUTE FORCE MATCHER OBJECT

```python
bfm = cv2.BFMatcher(cv2.NORM_L1,crossCheck=False)
```

# MATCHING

```python
match = bfm.match(train_des,test_des)
```

# SORTING

```python
match = sorted(match,key = lambda x : x.distance)
```

#  DRAWING MATCHING

```python
result =cv2.drawMatches(rgb, train_kp, rot_gray, test_kp, match, rot_gray, flags = 2)
```

# MATCHING DISPLAY

```python
plt.rcParams['figure.figsize']=[14.0, 10.0]
plt.title("BEST MATCHING")
plt.imshow(result)
```
